```python
#Step 1
from mysql.connector.pooling import MySQLConnectionPool
#Step 2
from mysql.connector import Error

try:
    #Step 3 - create a pool of connections
    conn_pool_a = MySQLConnectionPool(pool_name = "pool_a", pool_size=2, **dbconfig)
    print("pool of connections created")
    #Step 4 - get a connection from the pool and crate a cursor
    connection = conn_pool_a.get_connection()
    print("connection from the pool assigned")
    cursor = connection.cursor()
    print("cursor opened.")
except Error as err:
    print("ERROR:",err.errno, ":", err.msg)
```

```python
#Step 1 - Write a CREATE PROCEDURE query
create_proc_PeakHours = \
"""
CREATE PROCEDURE PeakHours()
SELECT HOUR(BookingSlot) as Hour, COUNT(*) AS BookingsQty
FROM Bookings
GROUP BY Hour
ORDER BY BookingsQty DESC;
"""
#Step 2 - Execute the CREATE PROCEDURE query
cursor.execute("DROP PROCEDURE IF EXISTS PeakHours;")
cursor.execute(create_proc_PeakHours)
print("procedure PeakHours created")
```

```python
#Step 3 - Call the stored procedure
cursor.reset()
cursor.callproc("PeakHours")

#Step 4 - Fetch the reusults after calling the procedure
result = next(cursor.stored_results())
dataset = result.fetchall()

#Step 5-6 - Extract and print names of the columns (headers)
h0, h1 = result.column_names
print(f"{h0:<6}{h1}")
print("-" * len(h0) + " " * (6-len(h0)) + "-" * len(h1))

#Step 7 - Print data records
for c0, c1 in dataset:
    print(f"{c0:<6}{c1}")
```

```python
#Step 1-4 - Write a CREATE PROCEDURE query
create_proc_GuestStatus = \
"""
CREATE PROCEDURE GuestStatus()
SELECT
CONCAT(b.GuestFirstName, ' ', b.GuestLastName) AS GuestFullName,
CASE
    WHEN e.Role in ('Manager','Assistant Manager') THEN 'Ready to pay'
    WHEN e.Role in ('Head Chef') THEN 'Ready to serve'
    WHEN e.Role in ('Assistant Chef') THEN 'Preparing Order'
    WHEN e.Role in ('Head Waiter') THEN 'Order served'
    ELSE 'Unknown'
END AS OrderStatus
FROM Bookings AS b
LEFT JOIN Employees AS e ON b.EmployeeID = e.EmployeeID
;
"""
#Step 5 - Execute the CREATE PROCEDURE query 
cursor.execute("DROP PROCEDURE IF EXISTS GuestStatus;")
cursor.execute(create_proc_GuestStatus)
print("procedure GuestStatus created.")
```

```python
#Step 6 - Call the stored procedure
cursor.reset()
cursor.callproc("GuestStatus")

#Step 7 - Fetch the reusults after calling the procedure
result = next(cursor.stored_results())
dataset = result.fetchall()

#Step 8-9 - Extract and print names of the columns (headers)
h0, h1 = result.column_names
print(f"{h0:<20}{h1}")
print("-" * len(h0) + " " * (20-len(h0)) + "-" * len(h1))

#Step 10 - Print data records
for c0, c1 in dataset:
    print(f"{c0:<20}{c1}")
```

```python
# Function sorted sorts comparing values in the first colum (FullName) by default
for c0, c1 in sorted(dataset):
    print(f"{c0:<20}{c1}")
```

```python
# Function sorted sorts comparing values in the 2nd column by specifying the key attribute as a simple lambda function.
for c0, c1 in sorted(dataset, key = lambda r: r[1]):
    print(f"{c0:<20}{c1}")
```

```python
# Function sorted sorts comparing values in the 2nd column by specifying the key attribute as a simple lambda function.
# The lambda function uses OrderStatusSort dictionary to return a numeric value assigned to each status

OrderStatusSort = {
    "Preparing Order": 1,
    "Ready to serve": 2,
    "Order served": 3,
    "Ready to pay": 4,
    "Unknown": 5
}
for c0, c1 in sorted(dataset, key = lambda r: OrderStatusSort[r[1]]):
    print(f"{c0:<20}{c1}")
```

```python
def get_connection_from_pool(conn_pool):
    """
    Get a connection for a guest from the given conn_pool.

    Parameters:
        conn_pool : MySQLConnectionPool
            a pool of connections where a connection for a guest should be taken from
    
    Returns:
        PooledMySQLConnections : if a connection was assigned to a guest
        None : if a connection could not be assigned
    """
    connection = None
    try:
        print("Getting a connection to the db from the pool ...")
        connection = conn_pool.get_connection()
    except Error as err:
        print("Could not get a connection from the pool!")
        print(err.errno, err.msg)
        try:
            print("The connection will be added manually using add_connection() ...")
            conn_pool.add_connection()
            connection = conn_pool.get_connection()
        except Error as err:
            print("New connection could not be added!")
            print(err.errno, err.msg)

    if isinstance(connection, PooledMySQLConnection):
        print("Successful connection to the database.")
    return connection


```

```python
def insert_booking(connection, new_booking):
    """
    Prepare and execute SQL INSERT query for a given new booking.
    
    Parameters:
        connection : PooledMySQLConnection
            to be used to execute the insert query
        new_booking : dict
            a dictionary with keys: TableNumber, FirstName, LastName, BookingTime, EmployeeID
            and corresponding values for a new booking
    """

    # prepare the insert query
    insert_query = \
    "INSERT INTO Bookings(TableNo, GuestFirstName, GuestLastname, BookingSlot, EmployeeID) VALUES " \
    "({}, \"{}\", \"{}\", \"{}\", {});".format(
    new_booking["TableNumber"],
    new_booking["FirstName"],
    new_booking["LastName"],
    new_booking["BookingTime"],
    new_booking["EmployeeID"]
    )

    GuestFullName = new_booking["FirstName"] + " " + new_booking["LastName"]
    print(f"Insert query : new booking for {GuestFullName}")
    print(insert_query)

    # execute the insert query
    try:
        if connection:  #the connection can be None is something went wrong
            cursor = connection.cursor()
            cursor.execute(insert_query)
            connection.commit()
            print(f"Added a new booking record for {GuestFullName}")
        else:
            print("No working connection!")
            print(f"Can't execute a booking query for {GuestFullName}")
    except Error as err:
        print(f"Problem with inserting a new booking record for for {GuestFullName}")
        print(err.errno, err.msg)
    except BaseException as err:
        print(f"Unexpected problem with inserting a new booking record for for {GuestFullName}")
        print(err)
```

```python
# New bookings will be added in a loop using the new_bookings dict defined earlier
# The keys of the new_bookings dict are 1,2,3

for guest_num in new_bookings:
    print(f"Adding a new booking record for a guest {guest_num} ...")
    # Get a database connection for a guest from the pool_b
    connections[guest_num] = get_connection_from_pool(conn_pool_b)    
    # insert to the database a new booking record
    insert_booking(connections[guest_num], new_bookings[guest_num])
    # print an empty separator line
    print()
```

```python
# Every connection stored in the connections container will be closed

print("Returning (closing) connections for Guests ...")
for guest_num, connection in connections.items():
    try:
        print(f"connection for a guest {guest_num} ... ", end="")
        if connection: connection.close()
        print("closed (returned).")
    except Error as err:
        print(f"can't be returned")
        print(err.errno, err.msg)
```
